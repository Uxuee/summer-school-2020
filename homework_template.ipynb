Graficar el campo eléctrico, las líneas de campo y el potencial para la siguiente configuración de cargas.

```python
# -*- coding: utf-8 -*-
"""
Created on Sat Feb  1 18:46:43 2020

@author: Ariadna
"""

import numpy as np
import matplotlib.pyplot as plt

xo1,yo1=0,0
xo2,yo2=0,10
xo3,yo3=10,0


def slope(xi,yi,q1,q2,q3,xo1,yo1,xo2,yo2,xo3,yo3):
    r1=((xi-xo1)**2+(yi-yo1)**2)**0.5
    r2=((xi-xo2)**2+(yi-yo2)**2)**0.5
    r3=((xi-xo3)**2+(yi-yo3)**2)**0.5
    
    Ex=q1*(xi-xo1)/(r1**3)+q2*(xi-xo2)/(r2**3)+q3*(xi-xo3)/(r3**3)
    Ey=q1*(yi-yo1)/(r1**3)+q2*(yi-yo2)/(r2**3)+q3*(yi-yo3)/(r3**3)

    return Ex,Ey

def potential(xi,yi,q1,q2,q3,xo1,yo1,xo2,yo2,xo3,yo3):
    r1=((xi-xo1)**2+(yi-yo1)**2)**0.5
    r2=((xi-xo2)**2+(yi-yo2)**2)**0.5
    r3=((xi-xo3)**2+(yi-yo3)**2)**0.5

    U=(q1/r1)+(q2/r2)+(q3/r3)
    
    return U
    
#la otra cosa
x=np.linspace(-8,15,40)
y=np.linspace(-8,15,40)
X,Y= np.meshgrid(x,y)

U,V=slope(X,Y,4e-6,3e-6,-7e-6,xo1,yo1,xo2,yo2,xo3,yo3)
P=potential(X,Y,4e-6,3e-6,-7e-6,xo1,yo1,xo2,yo2,xo3,yo3)

m_l=[(-5,-5),(0,5),(0,0),(-1,1),(-5,12),(6,6)]

plt.contour(X,Y,P,100)

plt.streamplot(X,Y,U,V) 
plt.scatter(xo1,yo1)
plt.scatter(xo2,yo2)
plt.scatter(xo3,yo3)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
